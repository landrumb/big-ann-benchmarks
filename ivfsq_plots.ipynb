```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colors as mcolors
import seaborn as sns
```

```python
data = pd.read_csv('res.csv')
data.head()
```

```python
parlayivf = data[data['algorithm'] == 'parlayivf']
faiss = data[data['algorithm'] == 'faiss']
parlayivf.head()
```

```python
# group parlayivf by dataset
parlayivf_grouped = parlayivf.groupby('dataset')
# parlayivf_grouped_wiki = parlayivf[parlayivf['dataset'].contains]
parlayivf_grouped.head()
```

```python
def pareto_curve(data):
    """ returns np arrays of x and y values for the pareto curve """
    data = data.sort_values('recall/ap', ascending=True)
    x = []
    y = []
    for index, row in data.iterrows():
        if len(y) == 0 or row['qps'] > y[-1]:
            x.append(row['recall/ap'])
            y.append(row['qps'])
            print(row['parameters'])
    return np.array(x), np.array(y)
```

```python
linestyles = {
    'wiki-10M' : {'color': 'black', 'marker': 'o', 'linestyle': '-', 'name': 'overall'},
    'wiki-10M-and-largest_0-20000': {'color': 'lightcoral', 'marker': 'o', 'linestyle': '-', 'name': 'small$\\times$small'},
    'wiki-10M-and-largest_20000+': {'color': 'firebrick', 'marker': 'o', 'linestyle': '--', 'name': '?$\\times$large'},
    'wiki-10M-and-smallest_0-20000': {'color': 'firebrick', 'marker': 'o', 'linestyle': '-.', 'name': 'small$\\times$?'},
    'wiki-10M-and-smallest_20000+': {'color': 'tomato', 'marker': 'o', 'linestyle': ':', 'name': 'large$\\times$large'},
    'wiki-10M-single_0-20000': {'color': 'indigo', 'marker': 'x', 'linestyle': '-', 'name':'small'},
    'wiki-10M-single_100000-400000': {'color': 'lightsteelblue', 'marker': 'x', 'linestyle': '--', 'name':'1st weight class'},
    'wiki-10M-single_20000-100000': {'color': 'cornflowerblue', 'marker': 'x', 'linestyle': '-.', 'name':'2nd weight class'},
    'wiki-10M-single_400000+': {'color': 'royalblue', 'marker': 'x', 'linestyle': ':', 'name':'3rd weight class'},
}
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
ax.set_xlabel('Recall')
ax.set_xlim(0, 1)
ax.set_ylabel('QPS')
ax.set_title('QPS vs Recall for IVF$^2$ on query regimes of wiki-10M')
for name, group in parlayivf_grouped:
    print(name)
    style = linestyles[name]
    x, y = pareto_curve(group)
    ax.plot(x, y, label=style['name'], color=style['color'], marker=style['marker'], linestyle=style['linestyle'], markersize=10)
ax.legend(loc='upper left')
plt.show()
```
